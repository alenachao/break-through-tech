```python
# analyze dataset
print(f"shape: {df.shape}")

unique_counts = df.apply(lambda x: len(x.unique()))
missing_counts = (len(df) - df.count())

summary_df = pd.DataFrame({
    'Unique Values': unique_counts,
    'Missing Values': missing_counts
})

summary_df
```

### Findings:
- We have 164 countries (164 classes)
- We already do not want to use year (most likely does not give new information), GINI index (many missing values).
- We will deprioritize confidence in national government, democratic quality, and delivery quality for similar reason.
- for all NANs we will impute the average

```python
# impute missing values
df = df.fillna(df.mean())
missing_counts = (len(df) - df.count())
missing_counts
```

```python
# visualize data
# to avoid overplotting lets do 15 countries
# we play around with features until we find some where the bars are very different
avg_social = df.groupby('country')['Social support'].mean()[:15]
avg_gdp = df.groupby('country')['Log GDP per capita'].mean()[:15]
avg_generosity = df.groupby('country')['Generosity'].mean()[:15]
avg_government = df.groupby('country')['Confidence in national government'].mean()[:15]

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

avg_social.plot(kind='bar', ax=axes[0, 0])
axes[0, 0].set_title('Average Social support by Country')
axes[0, 0].set_xlabel('Country')
axes[0, 0].set_ylabel('Average Social support')

avg_gdp.plot(kind='bar', ax=axes[0, 1])
axes[0, 1].set_title('Average Log GDP per capita by Country')
axes[0, 1].set_xlabel('Country')
axes[0, 1].set_ylabel('Log GDP per capita')

avg_generosity.plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Average Generosity by Country')
axes[1, 0].set_xlabel('Country')
axes[1, 0].set_ylabel('Generosity expectancy')

avg_government.plot(kind='bar', ax=axes[1, 1])
axes[1, 1].set_title('Average Positive affect by Country')
axes[1, 1].set_xlabel('Country')
axes[1, 1].set_ylabel('Positive affect expectancy')

plt.tight_layout()
plt.show()
```

```python
# YOUR CODE HERE
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan. 

You will:

1. Prepare your data for your model.
2. Fit your model to the training data and evaluate your model.
3. Improve your model's performance by performing model selection and/or feature selection techniques to find best model for your problem.

Add code cells below and populate the notebook with commentary, code, analyses, results, and figures as you see fit. 

```python
X = df[['Log GDP per capita','Healthy life expectancy at birth','Perceptions of corruption','Positive affect','Democratic Quality','Generosity']]
y = df['country']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
```
